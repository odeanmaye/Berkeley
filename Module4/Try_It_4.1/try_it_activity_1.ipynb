````python
import plotly.express as px
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
#Average yearly amount of wine purchased by income
df1 = df.copy()
df1['MntWinesYrAvg'] = (df1['MntWines']/2)

fig = px.scatter(df1[(df1['Income'] <= 200000)], x="Income", y="MntWinesYrAvg", labels={"Income": "Yearly Household Income", "MntWinesYrAvg": "Average Yearly Amount Spent On Wine"})

fig.write_image('./images/q1.png')
fig.show()
````

````python
#Yearly amount spent on wine as a share of yearly income
df1['MntWinesShareOfIncome'] = df1['MntWinesYrAvg'] / df1['Income']

fig = px.scatter(df1[(df1['Income'] <= 200000)], 
                 x="Income", 
                 y="MntWinesShareOfIncome", 
                 labels={
                     "Income": "Yearly Household Income", 
                     "MntWinesShareOfIncome": "Average Yearly Amount Spent On Wine<br>As A Share of Income"
                 })

fig.write_image('./images/q1_2.png')
fig.show()
````

````python
#What customers are more likely to participate in the last promotional campaign
# Selecting relevant columns for the pair plot
pairplot_cols = [
    'Income', 'MntWines', 'MntMeatProducts', 'Response'
]

# Filter the dataframe for the selected columns
pairplot_df = df[pairplot_cols]

# Create the pair plot
sns.pairplot(pairplot_df[(pairplot_df['Income'] <= 200000)], hue='Response', diag_kind='kde')
plt.savefig('./images/q2.png')
plt.show()
````

````python
#What customers are more likely to participate in the last promotional campaign
# Create a new column indicating whether a customer has children or not
df['Has_Children'] = (df['Kidhome'] + df['Teenhome']) > 0

# Calculate the average number of web purchases for customers with and without children
web_purchases_by_children = df.groupby('Has_Children')['NumWebPurchases'].mean().reset_index()
web_purchases_by_children.columns = ['Has_Children', 'Average NumWebPurchases']

# Map the boolean values to more readable labels
web_purchases_by_children['Has_Children'] = web_purchases_by_children['Has_Children'].map({True: 'Has Children', False: 'No Children'})

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Has_Children', y='Average NumWebPurchases', data=web_purchases_by_children)
plt.title('Average Number of Web Purchases by Customers with and without Children')
plt.ylabel('Average Number of Web Purchases')
plt.xlabel('Customer Type')
plt.savefig('./images/q3')
plt.show()
````

````python
# Define married statuses
married_statuses = ['Married', 'Together']

# Create a new column indicating whether a customer is married or in a partnership
df['Is_Married'] = df['Marital_Status'].isin(married_statuses)

# Calculate the average amount spent on wine for married and non-married customers
wine_purchases_by_marital_status = df.groupby('Is_Married')['MntWines'].mean().reset_index()
wine_purchases_by_marital_status.columns = ['Is_Married', 'Average MntWines']

# Map the boolean values to more readable labels
wine_purchases_by_marital_status['Is_Married'] = wine_purchases_by_marital_status['Is_Married'].map({True: 'Married/Together', False: 'Single'})

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='Is_Married', y='Average MntWines', data=wine_purchases_by_marital_status)
plt.title('Average Amount Spent on Wine by Marital Status')
plt.ylabel('Average Amount Spent on Wine')
plt.xlabel('Marital Status')
plt.savefig('./images/q4')
plt.show()
````

````python
def preferred_channel(row):
    channels = {
        'Web': row['NumWebPurchases'],
        'Catalog': row['NumCatalogPurchases'],
        'Store': row['NumStorePurchases']
    }
    preferred = max(channels, key=channels.get)
    return preferred

df['PreferredChannel'] = df.apply(preferred_channel, axis=1)

# Group the data by preferred channel and complaint status
channel_complaints = df.groupby(['PreferredChannel', 'Complain']).size().unstack(fill_value=0)
channel_complaints['Total'] = channel_complaints.sum(axis=1)
channel_complaints['Complaint Rate'] = channel_complaints[1] / channel_complaints['Total']

# Reset the index to turn the preferred channel into a column
channel_complaints = channel_complaints.reset_index()

# Plot the bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='PreferredChannel', y='Complaint Rate', data=channel_complaints)
plt.title('Complaint Rate by Preferred Purchasing Channel')
plt.ylabel('Complaint Rate')
plt.xlabel('Preferred Purchasing Channel')
plt.savefig('./images/q5.png')
plt.show()
````
