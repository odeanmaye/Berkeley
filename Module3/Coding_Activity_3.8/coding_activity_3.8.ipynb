### Problem 10: Customer Service Calls and Churn

**1 Point**

Use the `groupby` method to split the data by customers who had more than 4 customer service calls.  Use the `mean()` function to determine the average churn rate for each of the groups and assign your result as a DataFrame to `cust_service_churn` below.

````python
### GRADED

cust_service_churn = None

# YOUR CODE HERE
cust_service_churn = churn_df.groupby(churn_df['custserv_calls'] > 4)[['churn']].mean()

print(type(cust_service_churn))
cust_service_churn
````

It seems there are certain criteria where we find a larger percentage of customers churned based on this initial investigation.  Of course, there are many other ways to split, filter, and aggregate the data.  You are encouraged to further slice and dice the data to dig for more clues as to what drives customer churn.

````python
print('Median Daytime Minutes Charge: ' + str(churn_df['day_charge'].median()))
print('Median Evening Minutes Charge: ' + str(churn_df['eve_charge'].median()))
print('Median Nighttime Minutes Charge: ' + str(churn_df['night_charge'].median()))
print('Median International Minutes Charge: ' + str(churn_df['intl_charge'].median()))

#Calculate average total charges for each category
avg_day_charge = churn_df['day_charge'].median()
avg_eve_charge = churn_df['eve_charge'].median()
avg_night_charge = churn_df['night_charge'].median()
avg_intl_charge = churn_df['intl_charge'].median()

# Create a list of average charges
charges = [avg_day_charge, avg_eve_charge, avg_night_charge, avg_intl_charge]
labels = ['Daytime Charge', 'Evening Charge', 'Nighttime Charge', 'International Charge']

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(charges, labels=labels, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Average Total Charges')
plt.savefig('./out/total_charges.png')
plt.show()
````

````python
# Calculate the 85th percentile value for custserv_calls
percentile_85_value = churn_df['day_charge'].quantile(0.85)
print(percentile_85_value)

# Filter the dataframe to get the number of customers within the 85th percentile
customers_within_85th_percentile = churn_df[churn_df['day_charge'] <= percentile_85_value]

# Get the number of customers within the 85th percentile
num_customers_within_85th_percentile = customers_within_85th_percentile.shape[0]

# Get the number of customers
num_customers = churn_df.shape[0]

#Percent of customers in 85 percentile
print((num_customers_within_85th_percentile / num_customers) * 100)

sns.histplot(churn_df['day_charge'], kde=False, bins=10)
plt.xlabel('Amount of Daytime Charges')
plt.ylabel('Frequency')
plt.title('Daytime Charges')
plt.savefig('./out/day_charge.png')
plt.show()
````

````python
print('Median Customer Service Calls: ' + str(churn_df['custserv_calls'].median()))

# Calculate the 85th percentile value for custserv_calls
percentile_85_value = churn_df['custserv_calls'].quantile(0.85)

# Filter the dataframe to get the number of customers within the 85th percentile
customers_within_85th_percentile = churn_df[churn_df['custserv_calls'] <= percentile_85_value]

# Get the number of customers within the 85th percentile
num_customers_within_85th_percentile = customers_within_85th_percentile.shape[0]

# Get the number of customers
num_customers = churn_df.shape[0]

#Percent of customers in 85 percentile
print((num_customers_within_85th_percentile / num_customers) * 100)

sns.histplot(churn_df['custserv_calls'], kde=False, bins=10)
plt.xlabel('Number of Customer Service Calls')
plt.ylabel('Frequency')
plt.title('Customer Service Calls')
plt.savefig('./out/custserv_calls.png')
plt.show()
````
